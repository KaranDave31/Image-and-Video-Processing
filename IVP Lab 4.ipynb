```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
input = cv2.imread('pic1.jpg')
```

```python
cv2.imshow('Original', input)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gray_image = cv2.cvtColor(input,cv2.COLOR_BGR2GRAY)
```

```python
resized_image = cv2.resize(gray_image,(500,680))
```

```python
cv2.imshow('Resized',resized_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Task 1

```python
low_contrast_image = cv2.convertScaleAbs(resized_image, alpha=0.5, beta=0)
cv2.imshow('low_contrast_image', low_contrast_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
brightened_image = cv2.convertScaleAbs(resized_image, alpha=1.5, beta=50)
cv2.imshow('brightened_image', brightened_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
darkened_image = cv2.convertScaleAbs(resized_image, alpha=0.8, beta=-50)
cv2.imshow('darkened_image', darkened_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
fig, axes = plt.subplots(3, 2, figsize=(10, 10))

axes[0, 0].imshow(darkened_image, cmap='gray')
axes[0, 0].set_title('Dark Image')
axes[0, 0].axis('off')

axes[0, 1].hist(darkened_image.ravel(), 256, [0, 256])
axes[0, 1].set_title('Histogram of Dark Image')
axes[0, 1].set_xlabel('Gray Level')
axes[0, 1].set_ylabel('Count')

axes[1, 0].imshow(brightened_image, cmap='gray')
axes[1, 0].set_title('Bright Image')
axes[1, 0].axis('off')

axes[1, 1].hist(brightened_image.ravel(), 256, [0, 256])
axes[1, 1].set_title('Histogram of Bright Image')
axes[1, 1].set_xlabel('Gray Level')
axes[1, 1].set_ylabel('Count')

axes[2, 0].imshow(low_contrast_image, cmap='gray')
axes[2, 0].set_title('Low Contrast Image')
axes[2, 0].axis('off')

axes[2, 1].hist(low_contrast_image.ravel(), 256, [0, 256])
axes[2, 1].set_title('Histogram of Low Contrast Image')
axes[2, 1].set_xlabel('Gray Level')
axes[2, 1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
hist, bins = np.histogram(resized_image.ravel(), 256, [0, 256])
```

```python
hist
```

```python
bins
```

```python
N = resized_image.size
N
```

```python
pdf = hist / N
pdf
```

```python
cdf = pdf.cumsum()
cdf
```

```python
K = 256
equalized_cdf = np.round(cdf * (K - 1))
```

```python
equalized_hist, _ = np.histogram(equalized_cdf, 256, [0, 256])
```

```python
equalized_img = equalized_cdf[resized_image.ravel()].reshape(resized_image.shape)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(121)
plt.imshow(resized_image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(122)
plt.imshow(equalized_img, cmap='gray')
plt.title('Equalized Image')
plt.axis('off')

plt.show()

plt.figure(figsize=(10, 5))

plt.subplot(121)
plt.bar(range(256), hist)
plt.title('Original Histogram')
plt.xlabel('Gray Level')
plt.ylabel('Count')

plt.subplot(122)
plt.bar(range(256), equalized_hist)
plt.title('Equalized Histogram')
plt.xlabel('Gray Level')
plt.ylabel('Count')

plt.show()
```
